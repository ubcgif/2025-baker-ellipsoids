# Checks for magnetics

## Check for Oblate

```python
import numpy as np
import matplotlib.pyplot as plt
from functions import OblateEllipsoid, ProlateEllipsoid, TriaxialEllipsoid, ellipsoid_magnetics, ellipsoid_gravity


# Define two oblate ellipsoids where a is very close to b
a = 50.0
yaw, pitch = 0.0, 0.0
centre = (0, 0, 0)
ellipsoid_1 = OblateEllipsoid(a=a, b=a + 1e-1, yaw=yaw, pitch=pitch, centre=centre)
ellipsoid_2 = OblateEllipsoid(a=a, b=a + 1e-3, yaw=yaw, pitch=pitch, centre=centre)
susceptibility = 0.1

# Define a background field
magnitude, inclination, declination = 55_000, 0.0, 90.0
b0 = (magnitude, inclination, declination)

# Define observation points
easting = np.linspace(0, 2 * a, 151)
northing, upward = np.zeros_like(easting), np.zeros_like(easting)
coordinates = (easting, northing, upward)
# coordinate arrays are required to be 2d for now
coordinates = tuple(np.atleast_2d(c) for c in coordinates)

# Compute magnetic field of both ellipsoids
be_1, _, _ = ellipsoid_magnetics(
    coordinates, ellipsoid_1, susceptibility, b0, field="b"
)
be_2, _, _ = ellipsoid_magnetics(
    coordinates, ellipsoid_2, susceptibility, b0, field="b"
)

be_1 = be_1.ravel()
be_2 = be_2.ravel()


print("Be fields inside the ellipsoids:")
print(f"  - Ellipsoid 1: {be_1[0]}")
print(f"  - Ellipsoid 2: {be_2[0]}")

# Plot
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)
ax1.plot(easting, be_1)
ax2.plot(easting, be_2)
ax2.set_xlabel("Easting [m]")
ax1.set_ylabel("Be of ellipsoid 1 [nT]")
ax2.set_ylabel("Be of ellipsoid 2 [nT]")
plt.show()
```
